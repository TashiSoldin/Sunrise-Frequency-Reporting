```python
from datetime import datetime, date, time
import numpy as np
import pandas as pd
```

## Data ingestion

```python
file_path = '../test'
```

```python
df = pd.read_excel(f'{file_path}/A82 copy.xlsx')
```

```python
df.head()
```

```python
df['Colour'] = df['Last Event'].apply(
    lambda x: 'Blue' if x == 'Floor check' else 
                ('Green' if x == 'Floor check - Booking cargo' else 
               ('Pink' if x == 'Inbound Manifest' else 
                ('Purple' if x == 'Loaded for Delivery' else 'Red')))
)
```

```python
df.head()
```

```python
def coloured_row(row):
    if row['Colour'] == 'Blue':
        return ['background-color: blue']*len(row)  
    elif row['Colour'] == 'Green':
        return ['background-color: green']*len(row)  
    elif row['Colour'] == 'Pink':
        return ['background-color: pink']*len(row)
    elif row['Colour'] == 'Purple':
        return ['background-color: purple']*len(row)
    else:
        return ['background-color: white']*len(row)  
```

```python
coloured_df = df.style.apply(coloured_row, axis=1)
```

```python
coloured_df.to_excel('modified_dataframe.xlsx', index=False)
```
